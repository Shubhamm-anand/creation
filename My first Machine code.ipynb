```python
import pandas as pd 
```

```python
melbourne_file_path='D:\MSc_2nd_year\JUPYTER/melb_data.csv'
```

```python
melbourne_data = pd.read_csv(melbourne_file_path)
```

```python
melbourne_data.columns
```

```python
melbourne_data = melbourne_data.dropna(axis=0)
```

```python
y=melbourne_data.Price  #target prediction
```

```python
melbourne_features= ['Rooms','Bathroom','Landsize','Lattitude']
```

```python
X = melbourne_data[melbourne_features]
```

```python
X.describe()
```

```python
X.head()
```

# Model Building

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
melbourne_model=DecisionTreeRegressor(random_state=1)
```

```python
melbourne_model.fit(X,y) #fit model
```

```python
print("Shuboy model check time")
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are ")
print(melbourne_model.predict(X.head()))
```

```python
predicted_value=melbourne_model.predict(X)
import seaborn as sns
```
