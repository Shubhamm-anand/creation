```python
def oxygen_balance(formula):
    """Calculates the oxygen balance of a given chemical formula.

    Args:
        formula (str): The chemical formula of the compound.

    Returns:
        float: The oxygen balance of the compound.
    """

    element_counts = {}
    current_element = ""
    current_count = 1

    # Parse the formula and count the elements
    for char in formula:
        if char.isalpha():
            if current_element:
                element_counts[current_element] = element_counts.get(current_element, 0) + current_count
                current_count = 1
            current_element = char
        elif char.isdigit():
            current_count = int(char)
        else:
            raise ValueError("Invalid character in formula")

    if current_element:
        element_counts[current_element] = element_counts.get(current_element, 0) + current_count

    # Calculate the oxygen balance
    molecular_weight = 0
    oxygen_atoms = 0

    for element, count in element_counts.items():
        atomic_weight = {
            "C": 12.011, "H": 1.008, "N": 14.007, "O": 16.00, "Cl": 35.45, "Al": 26.98, "B": 10.81, "Ca": 40.07, "Cl": 35.45, "F": 18.99, "K": 39.09, "Na": 22.99
                        }.get(element,0)

        molecular_weight += atomic_weight*count
        if element == "O":
            oxygen_atoms = count

    oxygen_balance=(element_counts.get("C", 0))
   
    return oxygen_balance

if __name__ == "__main__":
    formula = input("Enter the chemical formula: ")
    balance = oxygen_balance(formula)
    print("Oxygen balance:", balance)
```
