```python
import pandas as pd
import rdkit
from rdkit import Chem
from rdkit.Chem import Descriptors, rdMolDescriptors

# Function to calculate molecular weight
def calculate_molecular_weight(smiles):
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol:
            return Descriptors.MolWt(mol)
        else:
            return "Invalid SMILES string"
    except Exception as e:
        return f"Error: {str(e)}"

# Function to convert SMILES to molecular formula
def smiles_to_formula(smiles):
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol:
            return rdMolDescriptors.CalcMolFormula(mol)
        else:
            return "Invalid SMILES string"
    except Exception as e:
        return f"Error: {str(e)}"

# Load Excel sheet and process SMILES strings
def process_excel(file_path):
    df = pd.read_excel(file_path)
    if 'SMILES' not in df.columns:
        print("Error: No 'SMILES' column found in the provided file.")
        return
    
    df['Molecular Weight'] = df['SMILES'].apply(calculate_molecular_weight)
    df['Molecular Formula'] = df['SMILES'].apply(smiles_to_formula)
    
    output_path = 'processed_valid_smiles.xlsx'
    df.to_excel(output_path, index=False)
    print(f"Results saved to {output_path}")

# Example usage
if __name__ == "__main__":
    file_path = 'valid_smiles.xlsx'
    process_excel(file_path)
```
